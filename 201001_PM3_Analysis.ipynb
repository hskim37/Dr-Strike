## Data Analysis

```python
%store -r publishersT
%store -r articleInfoT
```

```python
import pyLDAvis
import pyLDAvis.sklearn
pyLDAvis.enable_notebook()
```

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
articleContent = []
for p in articleInfoT:
    articles = articleInfoT[p]
    for a in articles:
#         articleContent.append(' '.join([x.strip() for x in articles[a]['content'].split()]))
        articleContent.append(articles[a]['content'])
```

```python
print(len(articleContent))
```

## 2. Convert to document-term matrix

Next, the raw documents are converted into document-term matrix (dtm), possibly as raw counts or TF-IDF form.

We're creating two different vectorizers, to notice the difference that the feature extraction makes in the process of modeling.

```python
tf_vectorizer = CountVectorizer(strip_accents = 'unicode',
                                stop_words = ['무단','전재','무단전재','및','재배포','금지'],
                                min_df = 0.15,
                                max_df = 0.5,
                                ngram_range = (1,3),
                               lowercase = False)
dtm_tf = tf_vectorizer.fit_transform(articleContent)
print(dtm_tf.shape)
```

```python
# print(tf_vectorizer.vocabulary)
tf_vectorizer.get_feature_names()[:]
```

```python
tfidf_vectorizer = TfidfVectorizer(**tf_vectorizer.get_params())
dtm_tfidf = tfidf_vectorizer.fit_transform(articleContent)
print(dtm_tfidf.shape)
```

## 3. Fit Latent Dirichlet Allocation models

We will fit two LDA models, each with 20 topics:

```python
# for TF DTM
lda_tf = LatentDirichletAllocation(n_components=10, random_state=0)
lda_tf.fit(dtm_tf)

# for TFIDF DTM
lda_tfidf = LatentDirichletAllocation(n_components=10, random_state=0)
lda_tfidf.fit(dtm_tfidf)
```

## 4. Visualizing the models with pyLDAvis

```python
pyLDAvis.sklearn.prepare(lda_tf, dtm_tf, tf_vectorizer)
```

```python
pyLDAvis.sklearn.prepare(lda_tfidf, dtm_tfidf, tfidf_vectorizer)
```
