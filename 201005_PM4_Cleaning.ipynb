### Clean and Restructure Data

```python
allArticles = [] # list of dicts

for p in articleInfoT:
    articles = articleInfoT[p]
    for a in articles:
        newArticleDict = dict()
        newArticleDict['link'] = a
        content = [x.strip() for x in articles[a]['content'].split('\n')] # split and stripped list
        paragraphList = []
        for c in content: # c is a paragraph
            if 'ⓒ' not in c and not (('무단전재' in c or '무단 전재' in c) and ('재배포금지' in c or '재배포 금지' in c)):
                sentenceList = []
                for s in c.split('.'): # s is a sentence
                    if len(s)>3 and '@' not in s and '=' not in s and '|' not in s:
                        if '입니다' not in s and '습니다' not in s and '캡처' not in s and '캡쳐' not in s:
                            wordList = []
                            for w in s.split(): # w is a word
                                w = w.strip()
                                if '기자' not in w and '연합뉴스' not in w and p not in w and '뉴스1' not in w and '뉴시스' not in w:
                                    if 'co' not in w and 'kr' not in w and 'com' not in w:
                                        wordList.append(w)
                            sentence = ' '.join(wordList)
                            sentenceList.append(sentence)
                paragraph = '. '.join(sentenceList)
                paragraphList.append(paragraph)
        newContent = ' '.join(paragraphList)
        
        for key in articles[a]:
            newArticleDict[key] = articles[a][key]
        newArticleDict['content'] = newContent.strip()
        newArticleDict['publisher'] = p
        allArticles.append(newArticleDict)
```
